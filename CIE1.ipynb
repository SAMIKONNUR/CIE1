```python

import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_csv('/content/mtcars.csv')
x =df['wt']
y=df['mpg']
plt.scatter(df['wt'],df['mpg'])
plt.title('scatter plot')
plt.xlabel('wt')
plt.ylabel('mpg')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_csv('/content/mtcars.csv')
auto_count=(df['am']==0).sum()
manual_count=(df['am']==1).sum()

plt.bar(['automatic','manual'],[auto_count,manual_count])
plt.title('manual automatic')
plt.xlabel('type of transmission ')
plt.ylabel('count')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name='mtcars.csv'
df=pd.read_csv('/content/mtcars.csv')
plt.boxplot(df['mpg'])
plt.title("Box Plot of MPG")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```
